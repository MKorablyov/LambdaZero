```python
import pandas as pd
summ = pd.read_csv('summary_table_v1.4.csv')
raw = pd.read_csv('drugcomb_data_v1.4.csv')
```

```python
single_line = summ[summ['cell_line_name'] == 'A549']
```

```python
single_line
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

mono = single_line[single_line['drug_col'].isna()]
comb = single_line[~single_line['drug_col'].isna()]

mono['type'] = 'mono'
comb['type'] = 'comb'

together = pd.concat((mono[['type', 'css']], comb[['type', 'css']]))


sns.displot(data=together, x='css', hue='type', palette='rocket')
```

```python
sns.kdeplot(data=together, x='css', hue='type', palette='rocket')
```
